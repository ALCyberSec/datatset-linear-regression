# The Powerball lottery and the winning conditions

Powerball is an American lottery game offered by 44 states. It is coordinated by the Multi-State Lottery Association (MUSL). Powerball's minimum advertised jackpot is 40$ million (annuity); Powerball's annuity is paid in 30 graduated installments or winners may choose a lump sum payment instead.

Drawings for Powerball are held every Wednesday and Saturday evening at 10:59 p.m. Eastern Time. Since October 7, 2015, the game has used a 5/69 (white balls) + 1/26 (Powerballs) matrix from which winning numbers are chHalogen, manufactured by Smartplay International of Edgewater Park, New Jersey.osen, resulting in odds of 1 in 292,201,338 of winning a jackpot per play. Each play costs  2,or 3 with the Power Play option.

The minimum Powerball bet is $2. In each game, players select five numbers from a set of 69 white balls and one number from 26 red Powerballs; the red ball number can be the same as one of the white balls. The drawing order of the five white balls is irrelevant; all tickets show the white ball numbers in ascending order. Players cannot use the drawn Powerball to match two of their white numbers, or vice versa. Players can select their own numbers, or have the terminal pseudorandomly select the numbers (called "quick pick", "easy pick", etc.).

The dataset we have here is showing data from October 2015 to June 2018.

```python
import numpy as np
import pandas as pd
import math
from scipy.special import comb
```

```python
#Calculate how many combinations of 5 numbers can be drawn from 69 unique numbers
combinations_five_denominator = comb(69,5).astype(int)
#Calculate how many ways the correct Powerball number can be chosen from 26 unique numbers
ball_combinations_denominator = comb(26,1).astype(int)
#setting the chances of selecting the correct powerball
ball_chances_numerator = 1
#identifying the number of unique balls numbers
unique_numbers = 69
#the number of draws each time
number_of_draws = 5
#showing the results of both combinations possibilities
combinations_five_denominator, ball_combinations_denominator
```

```python
def probability(rank_of_price, matching_number_of_balls, boolean_matching_powerball):
    #identifying the combination parameters for the number of balls chosen correctly to be drawn
    #from a sample size, drawn from total unique numbers of the balls
    not_correct_draws = number_of_draws - matching_number_of_balls
    samples_not_drawn = unique_numbers - number_of_draws
    prize_combination = comb(5,matching_number_of_balls).astype(int) * comb(samples_not_drawn, not_correct_draws).astype(int)
    #assumin the basic number of ball chances numerator as if the correct ball is not originally selcted
    ball_chances_numerator = 25
    #the boolean shall be yes or no
    if boolean_matching_powerball == 'yes':
        #in case if yes, it means i need to select the correct powerball
        ball_chances_numerator = 1     
    #the number of chances of getting the correct lottery requirment
    prize_numerator    = ball_chances_numerator * prize_combination
    #the total number of possibilities required for this lottery arrangement
    prize_denominator  = ball_combinations_denominator * combinations_five_denominator
    #making the final number as 1 of all possibilites
    final_probability_numerator = (prize_numerator/prize_numerator).astype(int)
    final_probability_denominator = (prize_denominator//prize_numerator).astype(int)
    print('The probabilty of getting {} prize is {} in {}'.format(rank_of_price, final_probability_numerator, final_probability_denominator))
```

```python
probability('first', 5, 'yes')
print("."*20)
probability('second', 5, 'no')
print("."*20)
probability('third', 4, 'yes')
print("."*20)
probability('fourth', 4, 'no')
print("."*20)
probability('fifth', 3, 'yes')
print("."*20)
probability('sixth', 3, 'no')
print("."*20)
probability('seventh', 2, 'yes')
print("."*20)
probability('eighth', 1, 'yes')
print("."*20)
probability('ninth', 0, 'yes')
```
